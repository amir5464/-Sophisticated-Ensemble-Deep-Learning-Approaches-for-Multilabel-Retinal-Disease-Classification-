### Loading and Preparing Retinal Disease Dataset

This code  demonstrates how to load and prepare the Retinal Disease dataset for machine learning tasks. The dataset consists of training, validation, and test sets, each with corresponding label CSV files.

#### Load Label CSV Files

- `TRAIN_CSV_DIR`: Specifies the directory path to the training set's label CSV file.
- `VAL_CSV_DIR`: Specifies the directory path to the validation set's label CSV file.
- `TEST_CSV_DIR`: Specifies the directory path to the test set's label CSV file.

- `train_labels = pd.read_csv(TRAIN_CSV_DIR)`: Loads the training set labels from the CSV file into a Pandas DataFrame.
- `val_labels = pd.read_csv(VAL_CSV_DIR)`: Loads the validation set labels.
- `test_labels = pd.read_csv(TEST_CSV_DIR)`: Loads the test set labels.

#### Add Image Path Column

- `add_image_path_column(df, directory)`: A custom function that adds an 'IMG_DIR' column to the DataFrame, which represents the full file path to each image.

- `directories`: A dictionary mapping dataset names to their corresponding image directories.

- A loop iterates through the datasets, calling `add_image_path_column` for each to add the 'IMG_DIR' column.

#### Concatenate Datasets

- `rfmid = pd.concat([train_labels, val_labels, test_labels], ignore_index=True)`: Concatenates the training, validation, and test label DataFrames into a single DataFrame, 'rfmid'.

- The result is a unified dataset ready for further preprocessing and machine learning tasks.



````python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
pd.options.display.max_columns = 50

TRAIN_CSV_DIR = "/kaggle/input/retinal-disease-classification/Training_Set/Training_Set/RFMiD_Training_Labels.csv"
VAL_CSV_DIR = '/kaggle/input/retinal-disease-classification/Evaluation_Set/Evaluation_Set/RFMiD_Validation_Labels.csv'
TEST_CSV_DIR='/kaggle/input/retinal-disease-classification/Test_Set/Test_Set/RFMiD_Testing_Labels.csv'

train_labels = pd.read_csv(TRAIN_CSV_DIR)
val_labels = pd.read_csv(VAL_CSV_DIR)
test_labels=pd.read_csv(TEST_CSV_DIR)

import os
import pandas as pd

def add_image_path_column(df, directory):
    df['IMG_DIR'] = df['ID'].apply(lambda image_id: os.path.join(directory, f"{image_id}.png"))


directories = {
    "train_labels": '/kaggle/input/retinal-disease-classification/Training_Set/Training_Set/Training',
    "val_labels"  : '/kaggle/input/retinal-disease-classification/Evaluation_Set/Evaluation_Set/Validation',
    "test_labels" : '/kaggle/input/retinal-disease-classification/Test_Set/Test_Set/Test'
}


for dataset_name, directory in directories.items():
    add_image_path_column(globals()[dataset_name], directory)
test_labels


rfmid = pd.concat([train_labels, val_labels, test_labels], ignore_index=True)
df=rfmid
df.head()
````

### Processing Disease Labels in a DataFrame

This code  processes disease labels in a DataFrame to keep the top 27 labels and create an "Other" label column.

#### Select Label Columns and Calculate Counts

- `label_columns = df.columns[2:-1]`: Selects the columns containing disease labels.
- `label_counts = df[label_columns].sum()`: Calculates the total count of each disease label.

#### Select Top 27 Labels

- `top_27_labels = label_counts.sort_values(ascending=False).head(27).index`: Selects the top 27 labels by count.

#### Create 'Other' Label Column

- A new column 'Other' is created and initialized to 0.

#### Iterate and Mark 'Other' Label

- The code iterates through rows and marks the 'Other' label if other labels are present.

#### Remove Unwanted Columns

- `columns_to_remove = [col for col in label_columns if col not in top_27_labels]`: Creates a list of columns to remove.
- The code removes the columns not in the top 27 labels.

#### Reorder Columns

- `column_order = list(df.columns[:-2]) + ['Other', 'IMG_DIR']`: Defines the desired column order.
- The columns are reordered according to the specified order.



````python
label_columns = df.columns[2:-1]
label_counts = df[label_columns].sum()

top_27_labels = label_counts.sort_values(ascending=False).head(27).index


df['Other'] = 0

for index, row in df.iterrows():
    if row[label_columns].sum() - row[top_27_labels].sum() > 0:
        df.at[index, 'Other'] = 1

columns_to_remove = [col for col in label_columns if col not in top_27_labels]


df = df.drop(columns=columns_to_remove)

column_order = list(df.columns[:-2]) + ['Other', 'IMG_DIR']
df = df[column_order]
df
````

### Generating a Pie Chart for Label Frequencies

This code  generates a pie chart to visualize the label frequencies in the dataset. It shows the distribution of labels as a percentage of the whole dataset.

#### Set a Threshold

- `threshold = 0.03`: Defines a threshold to filter out label frequencies below this threshold.

#### Calculate Label Frequencies

- `label_frequencies = label_counts / label_counts.sum()`: Calculates the label frequencies by dividing the count of each label by the total count of labels in the dataset.

#### Filter Labels

- `mask = label_frequencies > threshold`: Creates a mask to filter labels with frequencies above the threshold.
- `tail = label_frequencies.loc[~mask]`: Extracts labels with frequencies below the threshold.
- `label_frequencies = label_frequencies.loc[mask]`: Keeps labels above the threshold.
- `label_frequencies['Others'] = tail.sum()`: Combines the frequencies of labels below the threshold into an "Others" category.

#### Create a Pie Chart

- `plt.figure(figsize=(12, 8))`: Specifies the size of the pie chart.
- `label_frequencies.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)`: Generates a pie chart with labeled percentages, starting angle, and color scheme.
- `plt.ylabel("")`: Removes the y-axis label.




````python

threshold = 0.03


label_frequencies = label_counts / label_counts.sum()


mask = label_frequencies > threshold
tail = label_frequencies.loc[~mask]
label_frequencies = label_frequencies.loc[mask]
label_frequencies['Others'] = tail.sum()


plt.figure(figsize=(12, 8))
label_frequencies.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.ylabel("")

plt.savefig('/kaggle/working/Pie Chart.png', dpi=300)
plt.show()
````

### Exploratory Data Analysis: Correlation Heatmap

This code  performs Exploratory Data Analysis (EDA) by creating a correlation heatmap to visualize the relationships between different label columns in the dataset.

- `label_columns = df.columns[2:-1].tolist()`: Defines a list of label columns from the DataFrame.

#### Calculate Correlation Matrix

- `correlation_matrix = df[label_columns].corr()`: Calculates the correlation matrix between the label columns.

  - `ax=ax`: Specifies the axes to use for the heatmap.
  - `annot_kws={"size": 20, "color": "black"}`: Defines annotation settings for font size and color.
  - `cbar_kws={'label': 'Correlation Coefficient', 'shrink': 0.5}`: Adds a color bar with a label and adjusts its size.

- `ax.set_yticklabels(...)`: Sets the font size for the y-axis labels.

````python
import seaborn as sns

label_columns = df.columns[2:-1].tolist()
label_columns

correlation_matrix = df[label_columns].corr()


fig, ax = plt.subplots(figsize=(35, 30))


sns.heatmap(correlation_matrix, 
            cmap='coolwarm', 
            annot=True, 
            fmt=".2f", 
            linewidths=1, 
            ax=ax, 
            annot_kws={"size": 20, "color": "black"},
            cbar_kws={'label': 'Correlation Coefficient', 'shrink': 0.5})


ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=14)
ax.set_yticklabels(ax.get_yticklabels(), fontsize=14)

plt.tight_layout()

plt.savefig('/kaggle/working/EDA Correlation.png', dpi=300)
plt.show()
````

### Visualizing Sample Images for Each Label

This code  visualizes sample images associated with each label in the dataset. It provides an overview of images for specific labels.

#### Collect Sample Images

- `subset_image_paths = {}`: Initializes an empty dictionary to store sample image paths.

- Loop through each label in the dataset columns:
  - `for label in df.columns[2:-2]:`: Iterates through labels.

  - `sample = df[df[label] == 1].iloc[0]`: Finds the first sample in the dataset that is labeled with the current label.

  - `subset_image_paths[label] = sample['IMG_DIR']`: Stores the image path of the sample associated with the label.

#### Create Image Grid

- `grid_size = int(np.ceil(np.sqrt(len(subset_image_paths))))`: Calculates the grid size based on the number of labels to create a square grid of images.

- `fig, axes = plt.subplots(grid_size, grid_size, figsize=(15, 15))`: Sets up subplots with the specified grid size and figure size for displaying images.

- Loop through the image paths and labels:
  - `for ax, (label, img_path) in zip(axes.ravel(), subset_image_paths.items()):`: Iterates through the subplots and associated image paths.

    - `img = mpimg.imread(img_path)`: Reads the image from the image path.

    - `ax.imshow(img)`: Displays the image in the current subplot.

    - `ax.set_title(label)`: Sets the title of the subplot as the label.

    - `ax.axis('off')`: Removes axis labels and ticks for better visualization.

- Hide remaining empty subplots:
  - `for ax in axes.ravel()[len(subset_image_paths):]:`: Iterates through any remaining empty subplots.
    - `ax.axis('off')`: Hides these empty subplots.

#### Adjust Plot Layout and Save

- `plt.subplots_adjust(hspace=0.2)`: Adjusts the vertical spacing between subplots.

- `plt.tight_layout()`: Ensures that subplots are properly arranged within the figure.

- `plt.savefig("image_grid_tight.png", dpi=300)`: Saves the image grid as a file with high resolution.

- `plt.show()`: Displays the image grid.




````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


subset_image_paths = {}
for label in df.columns[2:-2]:  
    sample = df[df[label] == 1].iloc[0]
    subset_image_paths[label] = sample['IMG_DIR']


grid_size = int(np.ceil(np.sqrt(len(subset_image_paths))))

fig, axes = plt.subplots(grid_size, grid_size, figsize=(15, 15))
for ax, (label, img_path) in zip(axes.ravel(), subset_image_paths.items()):
    img = mpimg.imread(img_path)
    ax.imshow(img)
    ax.set_title(label)
    ax.axis('off')

for ax in axes.ravel()[len(subset_image_paths):]:
    ax.axis('off')

plt.subplots_adjust(hspace=0.2) 
plt.tight_layout()
plt.savefig("image_grid_tight.png", dpi=300) 
plt.show()
````
